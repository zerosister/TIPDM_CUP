**阶跃函数**
$$
u(t-t_0)=\begin{cases}
0\quad (t\lt t_0)\\
1\quad (t\gt t_0)
\end{cases}
$$

```python
# 阶跃信号
x = np.arange(100)
plt.figure(figsize=(3,3))
plt.plot(x, x>40)
plt.title('阶跃信号')
```

```python
plt.figure(figsize=(3,3))
plt.title('矩形脉冲')
x1 = x - 50 # 将时间轴平移 T/2
data1 = (x1>-50).astype(int)
data2 = (x1>=49).astype(int)
plt.plot(x1, data1-data2)
```

$$
\text{sgn}(t) = \begin{cases}
1 & \text{if } t \geq 0 \\  
-1 & \text{if } t < 0
\end{cases}
$$
$$
\text{sgn}(t) = 2u(t)-1
$$

```python
x=np.arange(-50,50)
plt.figure(figsize=(3,3))
data = (x>0).astype(int)
plt.plot(x, 2*data-1)
plt.grid()
plt.axhline(y=0, color='black', linewidth=2, linestyle='-.')
```

**单位冲激信号**

宽为 $\tau$，高为 $\cfrac{1}{\tau}$ 的矩形脉冲。使 $\tau$ 趋近于 0，则脉冲幅度 $\frac{1}{\tau}$ 趋近于无穷大。极限情况为 **单位冲激函数**，记为 $\delta(t)$。

$$
\delta(t) = \lim_{\tau \to 0} \frac{1}{\tau} [u(t+\tau/2)-u(t-\tau/2)]
$$

若矩形脉冲的面积不是 1 而是 $E$，则表示一个冲激强度为 $E$ 倍单位值冲激信号。 $E\delta(t)$

> 当然，冲激信号也可由其他波形推导出来，如正弦波、方波、三角波等。

Dirac 给出 $\delta(t)$ 的定义：
$$
\begin{cases}
\int_{-\infty}^{\infty} \delta(t) dt = 1 \\[1em]
\delta(t)=0 \text{  ,} t\neq 0
\end{cases}
$$

对于冲激函数有如下性质，设 $f(t)$ 在 $t=0$ 处连续，且处处有界，则：
$$
\begin{aligned}
\int_{-\infty}^{\infty} f(t) \delta(t) dt &= \int_{-\infty}^{\infty} f(0)\delta(t) dt \\[1em]
&= f(0)\int_{-\infty}^{\infty} \delta(t) dt \\[1em]
&= f(0)
\end{aligned}
$$
