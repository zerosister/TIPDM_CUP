## DataFrame basic operations

```python
# Series 
ser1 = pd.Series([1, 2, 'a'], index=['a', 'b', 'c'])
print(ser1, '\n')
ser2 = pd.Series({'a' : [1, 2, 3], 'b' : ['1.', '2.', '3.']})
print(ser2)
```

```python
# creating dataframe (DataFrame)
d = [[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
print(d)
df = pd.DataFrame(d, index=['a', 'b', 'c', 'd'], columns=['A', 'B', 'C', 'D']) # have the rows' and cols' headers
print(df)

# creating missing value dataframe
print('\n', pd.DataFrame(index=['1', '2'], columns=['a', 'b']))

# creating all-zeros dataframe
print(pd.DataFrame(0 ,index=['1', '2'], columns=['a', 'b']), '\n')
print(pd.DataFrame(1 ,index=['1', '2'], columns=['a', 'b']), '\n')

# creating dataframe using dictionary
d={'color':['blue','green','yellow','red','white'],
'object':['ball','pen','pencil','paper','mug'],
'price':[1.2,1.0,0.6,0.9,1.7]} # each key refers to a col
frame=pd.DataFrame(d,index=['a','b','c','d','e'])
print(frame)
```

### DataFrame attributes


```python
d = [[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
print(d)
df = pd.DataFrame(d, index=['a', 'b', 'c', 'd'], columns=['A', 'B', 'C', 'D']) # have the rows' and cols' headers
print(df)
print(df.values)
print(df.index)
print(df.columns)
print(df.dtypes)
print(df.size)
print(df.ndim)
print(df.shape)
```

### mutate elements

```python
# accessing the data
d = [[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df = pd.DataFrame(d, index=['a', 'b', 'c', 'd'], columns=['A', 'B', 'C', 'D']) # have the rows' and cols' headers
print(df)

print(df['A']) # a series, single col
print(df[['A', 'C']]) # multiple cols

print(df.head(3)) # .head(i) return data's first i rows
print(df.tail(2)) # .tail(j) return data's last j rows

# according to the location
print('\n', df.iloc[0, 0])
print(df.iloc[0:3, 2:4])
print(df.iloc[:, 2]) # col
print(df.iloc[0, :]) # row

# according to the name of row & col
print('\n', df.loc['a', 'A'])
print(df.loc['a':'c', 'C':'D']) # for names,it will be ['a', 'c'] not as int index ['a','c')

print(df.iloc[:, 0],' this is a Series')
print(df.iloc[:, 0:1], ' this is a dataFram')
print(type(df.iloc[:, 0]), ' ', type(df.iloc[:, 0:1]))
```

```python
# mutate dataframe
d = [[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df = pd.DataFrame(d, index=['a', 'b', 'c', 'd'], columns=['A', 'B', 'C', 'D']) # have the rows' and cols' headers
print(df)
df.loc['a', 'A'] = 1.093
print(df)

df.loc[:, 'B'] = [1.0, 2.0, 3.0, 4.0]
print(df)
```

```python
# adding elements
d = [[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df = pd.DataFrame(d, index=['a', 'b', 'c', 'd'], columns=['A', 'B', 'C', 'D']) # have the rows' and cols' headers
print(df)

df['E'] = 5
print(df)

df['F'] = [1, 2, 3, 4]
print(df)
```

```python
# delete elements(according to row or col)
d = [[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df = pd.DataFrame(d, index=['a', 'b', 'c', 'd'], columns=['A', 'B', 'C', 'D']) # have the rows' and cols' headers
print(df)

# df.drop('D') # default the axis is 0, error. the axis refers to the lable's axis
print(df.drop('D', axis=1))
print(df) # the drop won't do inplace by default
df.drop('D', axis=1, inplace=True)
print(df)
```

### describe DataFrame data


```python
import numpy as np
d = [[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df = pd.DataFrame(d, index=['a', 'b', 'c', 'd'], columns=['A', 'B', 'C', 'D']) # have the rows' and cols' headers
print(df)
print(np.mean(df, axis=1))
print(df.mean(axis=1))
print(df.std())
print(df.T.describe()) # dataframe after transpose
    
df['A'].value_counts()
```

## dealing with Time sequence data


```python
# TimeStamp
order = pd.read_csv('learning/data/meal_order_info.csv', encoding='gbk')
# print(order)
print(order['lock_time'].dtypes)
order['lock_time'] = pd.to_datetime(order['lock_time'])
print(order['lock_time'].dtypes)

print(pd.DatetimeIndex(order['lock_time']))
print(pd.PeriodIndex(order['lock_time'], freq='T')) # freqency is specified
# order['lock_time']
```

```python
print(order['lock_time'][0].year) # single value
print(order['lock_time'].dt.year)
print(order['lock_time'].dt.month)
```

```python
print(order['lock_time'] + pd.Timedelta(days=1))
print(order['lock_time'][1] - order['lock_time'][0])
```

## Divided into groups


```python
detail = pd.read_excel('learning/data/meal_order_detail.xlsx')
print(detail.head())

# divided by order_id
detail_group = detail[['order_id', 'counts', 'amounts']].groupby(by='order_id')
detail_group
```

### aggregate the data


```python
detail_group = detail[['order_id', 'counts', 'amounts']].groupby(by='order_id')
print(detail_group.agg('mean').head(3))
print(detail_group.agg(['mean', 'sum']).head(3))
print(detail_group.agg({'counts' : ['mean', np.max], 'amounts' : 'std'}).head(3)) # for 'counts' do 'mean' and 'np.max'
print(detail_group.agg({'counts' : lambda x: sum(x**2) }).head(3)) # to define a function
```

### Creating pivot_table ()


```python
detail = pd.read_excel('learning/data/meal_order_detail.xlsx')
detail.head()
```

```python
pd.pivot_table(detail[['order_id', 'counts', 'amounts']], index='order_id', aggfunc='sum').head(3) # get the sum of counts & amounts
```

```python
pd.pivot_table(detail[['order_id', 'dishes_name', 'counts']], index='order_id', columns='dishes_name' ,aggfunc='sum').head(3)
pd.pivot_table(detail[['order_id', 'dishes_name', 'counts']], index='order_id', columns='dishes_name', values='counts', fill_value=0).head(3)
```

### cross table

```python
pd.crosstab(index=detail['order_id'], columns=detail['dishes_name']).head(3)
```

```python
pd.crosstab(index=detail['order_id'], columns=detail['dishes_name'], values=detail['counts'], aggfunc='sum').fillna(0).head(3) # get the num of the dishes
```
