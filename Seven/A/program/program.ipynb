### 2.1 两产品线堆叠柱状图
分别按照月份对产品总数进行统计（产品总数为 ‘合格产品累计数’ + ‘不合格产品累计数’）

```python
M101_daily = M101.groupby(['月份', '日期']).agg({'合格产品累计数' : 'max', '不合格产品累计数' : 'max'})
M101_daily.reset_index(inplace=True)
M102_daily = M102.groupby(['月份', '日期']).agg({'合格产品累计数' : 'max', '不合格产品累计数' : 'max'})
M102_daily.reset_index(inplace=True)

M101_monthly = M101_daily.groupby('月份').agg({'合格产品累计数' : 'sum', '不合格产品累计数' : 'sum'})
M101_monthly['产品总数'] = M101_monthly['合格产品累计数'] + M101_monthly['不合格产品累计数']
M102_monthly = M102_daily.groupby('月份').agg({'合格产品累计数' : 'sum', '不合格产品累计数' : 'sum'})
M102_monthly['产品总数'] = M102_monthly['合格产品累计数'] + M102_monthly['不合格产品累计数']
```

```python
data1 = M101_monthly['产品总数']
data2 = M102_monthly['产品总数']

plt.figure(figsize=(10,6))
plt.bar(range(12), height=data1, label='M101')
plt.bar(range(12), height=data2, bottom=data1, label='M102')
for i in range(12):
    plt.text(i, y=data1.iloc[i]/2, s=f'{data1.iloc[i]:.2e}', va='bottom', ha='center')
    plt.text(i, y=data1.iloc[i]+data2.iloc[i]/2, s=f'{data2.iloc[i]:.2e}', va='bottom', ha='center')
plt.xlabel('月份')
plt.ylabel('产品总数（个）')
plt.title('各月产品总数')
plt.legend()
plt.xticks(ticks=range(12), labels=[f'{x+1} 月' for x in range(12)])
plt.savefig(os.path.join(image_dir, 'task2_1.png'))
plt.show()
```

## task 2.3 故障类别统计
发生故障的特征有：在一段连续的时间中，故障类别为同一种取值（非 None），由此可统计每天故障发生次数，进而统计全年发生总次数。

具体来说，在 Pandas 中这样实现：按照故障开始时列 ‘故障类别’ 的变化进行判断是否发生了故障
- 故障开始时，‘故障类别’会从 None 变为 A1/A2/A3/A4（分别对应 推出故障/抓取故障/安装故障/检测故障）

```python
M101_error = M101[M101['故障类别'].notna()]
pd.DataFrame(M101_error[['故障类别', '推出状态', '抓取状态', '安装状态', '检测状态']].value_counts())
```

```python
# r1 = M101['故障类别']
# r1.fillna('s', inplace=True)
# 假设没有两类不同的故障是连在一起的
# selected = [(r1.iloc[x-1] == 's') and (r1.iloc[x] != 's') for x in range(1, len(r1))]

r1 = M101['故障类别'].isna().tolist()
r2 = M102['故障类别'].isna().tolist()
# 错位异或，得到值为 1

# 假设没有两类不同的故障是连载一起的
# TODO
# check

selected = [((r1[x-1] == True) and (r1[x] == False)) for x in range(1, len(r1))] 
selected.insert(0, False) # 观察数据可知，第一个数据并未故障
sel_1 = pd.Series(selected)
print(sel_1.value_counts())

selected_2 = [((r2[x-1] == True) and (r2[x] == False)) for x in range(1, len(r2))] 
selected_2.insert(0, False) # 观察数据可知，第一个数据并未故障
sel_2 = pd.Series(selected_2)
print(sel_2.value_counts())
```

```python
M101['故障开始'] = sel_1
fault_start_1 = M101[sel_1]['故障类别']
counts_1 = pd.DataFrame(fault_start_1.value_counts().sort_index())

M102['故障开始'] = sel_2
fault_start_2 = M102[sel_2]['故障类别']
counts_2 = pd.DataFrame(fault_start_2.value_counts().sort_index())

print(counts_1)
print(counts_2)
labels = ['A1:推出故障', 'A2:抓取故障', 'A3:安装故障', 'A4:检测故障']
colors = ['lightblue', 'darkgreen', 'red', 'purple']

fig, ax = plt.subplots(figsize=(8,6))

# 绘制内层环形图（M101）
wedges1, texts1, autotexts1 = ax.pie(counts_1['count'],
       radius=0.7,
       autopct='%1.2f%%',
       colors=colors,
       pctdistance=0.55,
       wedgeprops=dict(width=0.3, edgecolor='w'))

# 绘制外层环形图（M102）
wedges2, texts2, autotexts2 = ax.pie(counts_2['count'],
       radius=1.0,
       autopct='%1.2f%%',
       colors=colors,
       pctdistance=1.2,
       wedgeprops=dict(width=0.3, edgecolor='w'))

ax.legend(wedges1, labels, title='故障类别', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

ax.text(0, 0, '内环：M101', ha='center', va='center', fontsize=12, fontweight='bold', color='gray')
ax.text(0, 1.1, '外环：M102', ha='center', va='center', fontsize=12, fontweight='bold', color='gray')

plt.title('两生产线各故障类别占比')
plt.savefig(os.path.join(image_dir, 'task2_3.png'))
```

## task2.4 故障发生持续时长
从 task2.3 中新增的‘故障开始’列中取得故障开始的时间，由取值得到生产线每 3s 一个行

```python
is_zero = pd.Series(M101_daily['合格产品累计数'] == 0)
print(is_zero.value_counts())

is_zero = pd.Series(M102_daily['合格产品累计数'] == 0)
is_zero.value_counts()
```

```python
# 采用 cumsum() 分类标签统计故障时间，由于连续的两个 NAN 会被识别为两个分组，故先 fillna
M101.fillna(0, inplace=True)
types_1 = M101['故障类别']
M101['tag'] = (types_1 != types_1.shift()).cumsum()
time_span_1 = M101.groupby(['故障类别', 'tag']).agg('size') * 3 # 注意每段时间是 3s
time_span_1 = time_span_1[time_span_1.index.get_level_values('故障类别') != 0]
time_span_1 = pd.DataFrame(time_span_1)
time_span_1.reset_index(inplace=True)
time_span_1.set_index('tag', inplace=True)
time_span_1.columns = ['故障类别', '持续时间']

labels = ['A1:推出故障', 'A2:抓取故障', 'A3:安装故障', 'A4:检测故障']
colors = ['yellow', 'darkgreen', 'red', 'purple']
cnt = 0
for ty in time_span_1['故障类别'].unique():
    data = time_span_1[time_span_1['故障类别'] == ty].copy()
    data.drop(columns=['故障类别'], inplace=True)
    series = data['持续时间']
    plt.hist(series, bins=30, alpha=0.5, label=labels[cnt], color=colors[cnt], edgecolor='black')
    cnt += 1
plt.legend()
plt.ylabel('频次')
plt.xlabel('持续时间（单位：s）')
plt.title('M101 全年故障发生持续时间直方图')
plt.savefig(os.path.join(image_dir, 'task2_4_1.png'))
```

首先筛选 4 月 26 日前 100s 生产线数据

```python
bools = (M101['月份']==4) & (M101['日期']==26) & (M101['时间'] < 34)
data = M101[bools][['时间','推出状态', '抓取状态', '安装状态', '检测状态']]
data
```

```python
bools = (M101['月份']==4) & (M101['日期']==26) & (M101['时间'] < 34)
data = M101[bools][['时间','推出状态', '抓取状态', '安装状态', '检测状态']]

results = []
for col in data.columns[1:]:
    pro = data[col]
    check = pd.DataFrame(data[['时间', col]])
    check['tag_'+col] = (pro != pro.shift()).cumsum()
    
    # 筛选掉非工作状态
    selected = (pro==1)
    
    check.columns = ['时间', col, 'tag_'+col]
    check.drop(columns=[col], inplace=True)
    check = check[selected]
    answer = check.groupby('tag_'+col).agg({'时间':['min', 'max']})
    result = [(answer.iloc[x, 0]*3-3, (answer.iloc[x, 1] - answer.iloc[x,0] + 1)*3) for x in range(len(answer))]
    results.append(result)

tasks = ['推出', '抓取', '安装', '检测']

# 为每个任务分配不同的颜色
colors = ['skyblue', 'lightgreen', 'salmon', 'orange']

# 每个任务的时间段 
time_spans = {
    '推出':results[0],
    '抓取':results[1], 
    '安装':results[2],
    '检测':results[3]
}

fig, gnt = plt.subplots(figsize=(10,5))

# 设置 Y 轴
gnt.set_yticks([10, 20, 30, 40])  # 每个任务的 y 值
gnt.set_yticklabels(tasks)     # 设置任务名称
gnt.set_ylim(5, 45)            # y 轴范围

# 设置 X 轴
gnt.set_xlabel('时间')
gnt.set_xlim(0, 100)

# 添加每个任务的时间跨度条
for i, task in enumerate(tasks):
    gnt.broken_barh(time_spans[task], (10 * (i + 1) - 5, 8), facecolors=colors[i])

# 显示图形
plt.title('4月26日 M101 前 100s 工序甘特图')
plt.ylabel('工序')
plt.show()
plt.savefig(os.path.join(image_dir, 'task2_5.png'))
```
