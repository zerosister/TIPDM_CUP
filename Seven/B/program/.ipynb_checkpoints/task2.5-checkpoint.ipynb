```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
###读取2的数据



data2_excel = pd.read_excel('Desktop/result2.xlsx')
```

```python
sui_cro = data2_excel['适用人群'].value_counts() 




###生成词频字典
wor_fre = sui_cro.to_dict()


###画词云图
wordcloud = WordCloud(width=1600, height=800, background_color='white',max_font_size=100,min_font_size=10)
wordcloud.generate_from_frequencies(wor_fre)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # 关闭坐标轴
plt.show()
```
