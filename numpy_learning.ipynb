```python
# array indexing
arr1 = np.array([0.1, 0.32, 0.3, 233.2, 3])
print(arr1)
print(arr1[0])
print(arr1[-5])

print(arr1[1:3])  # slice of array
print(arr1[-4:-2])

resl = arr1[3]
resl2 = arr1[3:4]
print(resl, type(resl), resl.shape)
print(resl2, type(resl2),  resl2.shape)
```

```python
# bool indexing
arr2 = np.array([2.3, 1.8, 4.5])
print(arr2)
print(arr2[[False, False, True]])
# use expression instead
index = arr2 > 2
print(index)
print(arr2[index])
```

```python
# multiple dimension indexing
arr3 = np.arange(1, 13).reshape(3, 4)
print(arr3)
print(arr3[2, 3], '\n')
print(arr3[2, :],'\n') # get whole row
print(arr3[:, 0],'\n') # get whole col
print(arr3[1:3, 1:3], '\n')

print(arr3[2:, :], "\n") # two slices get 2D array

print(arr3[arr3[:, 0] > 4, :])
arr3[:, 0] > 4
```

```python
# mutate elements
arr3 = np.arange(1, 13).reshape([3,4])
print(arr3)
arr3[0, 0] = 15
print(arr3)
```

```python
# calculate distance matrix
n = 100  # sample nums
x = np.linspace(1, 100, n)  # sample's row coordinate
y = np.linspace(1, 100, n)  # sample's col coordinate
print(f"x={x}\ny={y}")

dist = np.zeros([n, n])
for i in range(n):
    for j in range(n):
        dist[i, j] = np.sqrt((x[i] - x[j])**2 + (y[i] - y[j])**2) # calculate distance

print(dist)
```

### reshaping

```python
# reshaping
arr4 = np.arange(1, 13)
print(arr4,'\n\n', arr4.reshape(3, 4),'\n\n', arr4.reshape(4,3))

# flatten
arr5 = arr4.reshape(3,4)
print(arr5,"\n\n")
print(arr5.ravel())
print(arr5.flatten()) # flatten by cols
print(arr5.flatten('F'),'\n\n') # flatten by rows

# contact array
print(arr5,"\n")
print(np.hstack([arr5, arr5]), "\n") # (A, B) input as a tuple/list, hstack contact by cols
print(np.vstack((arr5, arr5)), "\n")
print(np.concatenate((arr5, arr5)), "\n")
print(np.concatenate((arr5, arr5), axis = 1), "\n")

# split array
print(np.hsplit(arr5, 2), "\n") # split array by cols
print(np.vsplit(arr5, 3), "\n") # split array by rows, second param is the num of result arrays 
print(np.split(arr5, 3), "\n")
print(np.split(arr5, 2, axis=1), "\n")
```

## Numpy matrix & functions

```python
import numpy as np
matr1 = np.mat("1 2 3;4 5 6;7 8 9")
print(matr1, type(matr1))

matr2 = np.matrix([[1, 2, 3],[4, 5, 6],[7, 8, 9]])
print(matr2)

# to cat more than 1 matrix into 1 matrix
print(np.bmat("arr1 arr2; arr1 arr2")) # don't understand

# basic matrix attributes
print("\n\n", matr1.T) # transpose
print("\n", matr1.I) # reverse
```

### Universial Function


```python
arr1 = np.array([0.2, 0.4, 0.6])
arr2 = np.array([1.2, 0.43, 23.6])
list1 = [.2, .4, .6]

print(arr1 + 1)
# print(list1 + 1) # illegal operation
print(arr1 * 2)
print(list1 * 2) # list will return a list with double lenth

print(arr1 + arr2) # elementwise add
print(arr1 * arr2) # elemetwise multiply
print(arr1 > arr2)

print(np.any(arr1 == 0.2))  # logic 'or'
print(np.all(arr1 == 0.2))  # logic 'and'
```

### BroadCasting

```python
arr3 = np.arange(1, 13).reshape(4, 3)
arr4 = np.arange(1, 4)
print(arr3)
print(arr4)
print(arr3 + arr4)

arr5 = np.arange(1, 5).reshape(4, 1)
print(arr3 + arr5)
```

## Numpy statistics analysis

reading & writing numpy(bits) files
* save: save bits data
* load
* savez: save multiple arrays into one file

```python
import numpy as np
# help(np.save)
arr1 = np.arange(1, 13).reshape(3, 4)
print(arr1)

np.save('tmp/arr1.npy', arr1) # save data
arr1_load = np.load('tmp/arr1.npy')
print(arr1_load)

arr2 = np.arange(1, 13).reshape(4, 3)

np.savez('tmp/arr1_2.npz', arr1, arr2) # save more than one array
arr1_2_load = np.load('tmp/arr1_2.npz')
print(arr1_2_load.files) # look out array's auto name
print(arr1_2_load['arr_0']) # use dict to access data
```

reading & writing txt files


```python
np.savetxt('tmp/arr1.txt', arr1, fmt="%.4f", delimiter=',') # fmt is format, delimiter is to divide data

# load data 
# arr1_load = np.loadtxt('tmp/arr1.txt') # this will go wrong for not specify delimiter
arr1_load = np.loadtxt('tmp/arr1.txt', delimiter=',')
print(arr1_load)
```

### Numpy statistics easy


```python
arr3 = np.random.randint(1, 10, (3, 4))
print(arr3)

arr3.sort(axis=0)  # sort rows
print(arr3)
print(arr3.argsort(axis=0)) # return sorted index

print(np.tile(arr3, 2)) # copy the whole array
print(np.repeat(arr3, 2, axis=1)) # copy elements

print(arr3.mean()) # the average of the whole array
print(arr3.mean(axis=0)) # mean according to the rows

print(arr3.argmax())
```
