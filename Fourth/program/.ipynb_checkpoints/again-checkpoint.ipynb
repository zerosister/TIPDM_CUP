```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = 'SimHei' # 设置中文显示
plt.rcParams['axes.unicode_minus'] = False
```

```python
project_path = os.path.join(os.getcwd(), os.pardir)
data_path = os.path.join(project_path, 'data.xlsx')
result_path = os.path.join(project_path, 'result/')
salesData = pd.read_excel(data_path, sheet_name='SalesData')
```

```python
# 数据预处理，缺失值，异常值，重复值
print(salesData.isna().sum())
print()
print(salesData.duplicated().value_counts())
```

## task 1.1 同比增长率
统计各个年度/季度中，地区、国家、服务分类的销售额和利润数据，并计算各国、各服务分类销售额和利润的同比增长率。

```python
salesData
```

```python
salesData['年度'] = salesData['日期'].dt.year
salesData['季度'] = salesData['日期'].dt.quarter
salesData
```

```python
country_year = salesData.groupby(by=['年度', '国家']).agg({'销售额': 'sum', '利润' : "sum"}).reset_index()
# 数据完整不空缺，共 52 个国家，4 年数据每个国家均有
# len(country_year.index.value_counts()) / 52 
country_year['销售额年增长率'] = country_year['销售额'].pct_change(periods=52) * 100
country_year['利润年增长率'] = country_year['利润'].pct_change(periods=52) * 100
country_year.fillna(0, inplace=True)

# 2020 年 销售额前三的国家及其年增长率
country_year[country_year['年度']==2020].sort_values('销售额', ascending=False)[['年度','国家', '销售额', '销售额年增长率']].head(3)
```

## task 2.2 同比增长率图表

```python
df = salesData.groupby(by=['年度', '服务分类']).agg({'销售额': 'sum'}).reset_index()
df['销售额年增长率'] = df['销售额'].pct_change(periods=3) * 100
df.fillna(0, inplace=True)

df
```

```python
plt.figure(figsize=(8,6))
plt.subplot()
bar_width = 0.2
serves = df['服务分类'].unique()
serve_2018 = df.loc[df['年度']==2018, '销售额年增长率']
serve_2019 = df.loc[df['年度']==2019, '销售额年增长率']
serve_2020 = df.loc[df['年度']==2020, '销售额年增长率']
plt.bar(np.arange(len(serves)), height=serve_2018, width=bar_width, label='2018', color='steelblue')
plt.bar(np.arange(len(serves))+bar_width, height=serve_2019, width=bar_width, label='2019', color='red')
plt.bar(np.arange(len(serves))+2*bar_width, height=serve_2020, width=bar_width, label='2020', color='purple')
plt.xticks(np.arange(len(serves)) + bar_width, serves)
plt.ylabel('销售额年增长率')
plt.title('2018~2020年度各服务分类销售额年增长率')
plt.xlabel('服务分类')
plt.legend()
```

## task 1.4 预测各个地区 2021 第一季度利润

```python
pre_df = salesData.groupby(by=['年度', '季度', '地区']).agg({'利润' : "sum"}).reset_index()
pre_df
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# 按照地区分别进行预测
for area in pre_df['地区'].unique():
    area_df = pre_df[pre_df['地区']==area]
    area_df.pop('地区')
    add_2021 = pd.DataFrame({'年度': [2021], '季度' : [1], '利润' : [0]})
    area_df = pd.concat([area_df, add_2021])
    # print(area_df)
    data = area_df[['年度', '季度']].to_numpy()
    scaled_train = MinMaxScaler().fit_transform(data)
    label = area_df['利润'].to_numpy()
    model = LinearRegression()
    model.fit(data[:-1], label[:-1])

    print(f"截距项：{model.intercept_}")
    print(f"特征 X1 的系数：{model.coef_[0]}")
    print(f"特征 X2 的系数：{model.coef_[1]}")

    y_pred = model.predict([data[-1]])
    print(f'{area}\'s 2021 1 quarter profit will be {y_pred}')
    print()
```
