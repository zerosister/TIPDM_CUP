```python
# 按照国家分类
# country_SD = salesData[['日期', '国家', '地区', '服务分类', '销售额', '利润']].groupby([salesData['日期'].dt.year, '国家'])['销售额'].sum().to_frame()
# country_SD_2 = salesData[['日期', '国家', '地区', '服务分类', '销售额', '利润']].groupby([salesData['日期'].dt.year, '国家'])['利润'].sum().to_frame()
country_SD_3 = salesData[['日期', '国家', '销售额', '利润']].groupby([salesData['日期'].dt.year, '国家']).agg({'销售额': 'sum', '利润' : 'sum'})
country_SD_3.head(3)
```

```python
pivoted_data = country_SD_3.pivot_table(values=['销售额', '利润'], columns=['日期'], index='国家')

pivoted_data.columns = pivoted_data.columns.rename(['指标', '日期'])
# 获取列名（年份）
years = pivoted_data.columns

# 创建新的 MultiIndex 列，用于存放增长率
new_columns = pd.MultiIndex.from_product(
    [['销售额年增长率', '利润年增长率'], pivoted_data.columns.levels[1]],
    names=['指标', '日期']
)

# 初始化增长率数据框，与 pivoted_data 具有相同的行索引和新创建的增长率列
growth_rates = pd.DataFrame(index=pivoted_data.index, columns=new_columns)

# 计算增长率并存入 growth_rates 数据框
for country in pivoted_data.index:
    for year in pivoted_data.columns.levels[1][1:]:  # 从第二年开始计算增长率
        # 获取销售额和利润的当前年和上一年的数据
        current_sales = pivoted_data.loc[country, ('销售额', year)]
        current_profit = pivoted_data.loc[country, ('利润', year)]
        prev_sales = pivoted_data.loc[country, ('销售额', year - 1)]
        prev_profit = pivoted_data.loc[country, ('利润', year - 1)]
        
        # 计算增长率
        sales_growth = (current_sales - prev_sales) / prev_sales * 100 if prev_sales != 0 else None
        profit_growth = (current_profit - prev_profit) / prev_profit * 100 if prev_profit != 0 else None
        
        # 将增长率存入相应位置
        growth_rates.loc[country, ('销售额年增长率', year)] = sales_growth
        growth_rates.loc[country, ('利润年增长率', year)] = profit_growth

# 将增长率数据合并到原始数据框中
pivoted_data = pd.concat([pivoted_data, growth_rates], axis=1)
```

## task 2.2 绘制销售额，利润年增长率和各季度同比增长率

```python
year_country = salesData[['日期', '国家', '服务分类', '销售额', '利润']].groupby([salesData['日期'].dt.year, '国家', '服务分类']).agg({'销售额': 'sum', '利润' : 'sum'})
year_region = salesData[['日期', '地区', '服务分类', '销售额', '利润']].groupby([salesData['日期'].dt.year, '地区', '服务分类']).agg({'销售额': 'sum', '利润' : 'sum'})
salesData_2 = salesData[['日期', '国家', '地区', '服务分类', '销售额', '利润']].copy()
salesData_2['年份'] = salesData_2['日期'].dt.year
salesData_2['月份'] = salesData_2['日期'].dt.month
salesData_2.pop('日期')
print(salesData_2)
season_country = salesData_2.groupby([salesData_2['年份'], salesData_2['月份'], '国家', '服务分类']).agg({'销售额': 'sum', '利润' : 'sum'}).reset_index()
season_region = salesData_2.groupby([salesData_2['年份'], salesData_2['月份'], '地区', '服务分类']).agg({'销售额': 'sum', '利润' : 'sum'}).reset_index()
```

```python
# 创建三个月的分组标签
season_country['季度'] = (season_country['月份'] - 1) // 3 + 1
season_region['季度'] = (season_region['月份'] - 1) // 3 + 1
season_country.pop('月份')
season_region.pop('月份')

# 进行聚合，每个国家/地区按季度汇总
season_country_quarterly = season_country.groupby(['年份', '季度', '国家', '服务分类']).agg({'销售额': 'sum', '利润': 'sum'})
season_region_quarterly = season_region.groupby(['年份', '季度', '地区', '服务分类']).agg({'销售额': 'sum', '利润': 'sum'})
season_country_quarterly
```

```python
def ratios(tab, cnt, indexs):
    # 初始化增长率数据框，与 tab 具有相同的行索引和新创建的增长率列
    growth_rates = pd.DataFrame(index=tab.index, columns=new_columns)
    
    for country_or_region in indexs.levels[0]:
        for year in tab.columns.levels[1][1:]: # 从第二年开始计算增长率
            for ty in indexs.levels[1]: # 三种不同种类
                current_sales = tab.loc[(country_or_region, ty), ('销售额', year)]
                current_profit = tab.loc[(country_or_region, ty), ('利润', year)]
                prev_sales = tab.loc[(country_or_region, ty), ('销售额', year - 1)]
                prev_profit = tab.loc[(country_or_region, ty), ('利润', year - 1)]

                # 计算增长率
                sales_growth = (current_sales - prev_sales) / prev_sales * 100 if prev_sales != 0 else None
                profit_growth = (current_profit - prev_profit) / prev_profit * 100 if prev_profit != 0 else None
                
                # 将增长率存入相应位置
                growth_rates.loc[(country_or_region, ty), ('销售额增长率', year)] = sales_growth
                growth_rates.loc[(country_or_region, ty), ('利润增长率', year)] = profit_growth

    # tab = pd.concat([tab, growth_rates], axis=1).dropna(axis=1)
    tab.to_csv(os.path.join(result_path, 'task1_1_%d.csv'%cnt))
    tabs.append(tab)
    print('task1_1_%d.csv'%cnt, ' saved')
```

```python
# 计算响应的同比增长率
dfs = [year_country, year_region]
# season_country_quarterly, season_region_quarterly
tabs = []
cnt = 1
for file in dfs:
    tab = file.pivot_table(index=file.index.names[1:3], values=['销售额', '利润'], columns='日期')
    new_columns = pd.MultiIndex.from_product(
        [['销售额增长率', '利润增长率'], tab.columns.levels[1]],
        names=['指标', '日期']
    )

    indexs = tab.index
    ratios(tab, cnt, indexs)
    cnt += 1
```
